```python
states=maindf.groupby('State name').size()
print(states)
```

```python
maindf.groupby('State name').size().plot.bar(color='pink')
plt.xlabel("States",size=20)
plt.ylabel("Districts_counts",size=20)
plt.title("District counts of the each States",fontsize=20)
plt.xticks(size=15)
plt.show()
```

```python
#Calculating state wise literacy rates
states=maindf.groupby(by="State name")
literacy_rate=[]
for value,group in states:
    total_state_pop=0
    total_literate_pop=0
    for row in group.iterrows():
        total_state_pop=total_state_pop+row[1][3]
        total_literate_pop=total_literate_pop+row[1][6]
    rate=(total_literate_pop/total_state_pop)*100
    literacy_rate.append((value,rate))
print (literacy_rate)
```

```python
literacy=pd.DataFrame(literacy_rate,columns=['State','Literacy rate'])
print(literacy)
```

```python
plt.figure(figsize=(30,20))
State_Name=list(literacy["State"])
plt.bar(literacy['State'],literacy['Literacy rate'],color='green')
plt.xticks(State_Name,rotation='90',Fontsize=30)
plt.yticks(Fontsize=30)
plt.title("State Literacy Rates",fontsize=30)
plt.show()
```

```python
mainwater=pd.DataFrame(columns=['State','Un_covered_well','Handpump_Tubewell_Borewell','water_Spring','River_Canal','Other_source',
                               'Spring_River_Canal_Tank_Pond_Lake_Other_sources','Tank_Pond_Lake','Tapwater','Tubewell_Borehole'])
mainwater['State']=maindf['State name']
mainwater['Un_covered_well']=maindf['Main_source_of_drinking_water_Un_covered_well_Households']
mainwater['Handpump_Tubewell_Borewell']=maindf['Main_source_of_drinking_water_Handpump_Tubewell_Borewell_Households']
mainwater['water_Spring']=maindf['Main_source_of_drinking_water_Spring_Households']
mainwater['River_Canal']=maindf['Main_source_of_drinking_water_River_Canal_Households']
mainwater['Other_source']=maindf['Main_source_of_drinking_water_Other_sources_Households']
mainwater['Spring_River_Canal_Tank_Pond_Lake_Other_sources']=maindf['Main_source_of_drinking_water_Other_sources_Spring_River_Canal_Tank_Pond_Lake_Other_sources__Households']
mainwater['Tank_Pond_Lake']=maindf['Main_source_of_drinking_water_Tank_Pond_Lake_Households']
mainwater['Tapwater']=maindf['Main_source_of_drinking_water_Tapwater_Households']
mainwater['Tubewell_Borehole']=maindf['Main_source_of_drinking_water_Tubewell_Borehole_Households']
a=mainwater['Un_covered_well'].sum()/1000000
b=mainwater['Handpump_Tubewell_Borewell'].sum()/1000000
c=mainwater['water_Spring'].sum()/1000000
d=mainwater['River_Canal'].sum()/1000000
e=mainwater['Other_source'].sum()/1000000
f=mainwater['Spring_River_Canal_Tank_Pond_Lake_Other_sources'].sum()/1000000
g=mainwater['Tank_Pond_Lake'].sum()/1000000
h=mainwater['Tapwater'].sum()/1000000
i=mainwater['Tubewell_Borehole'].sum()/1000000
x=[a,b,c,d,e,f,g,h,i]
lbl=['Un_covered_well','Handpump_Tubewell_Borewell','water_Spring','River_Canal','Other_source','Spring_River_Canal_Tank_Pond_Lake_Other_sources','Tank_Pond_Lake','Tapwater','Tubewell_Borehole']
plt.pie(x,labels=lbl,autopct='%1.1f%%')
plt.title("Main sources of Drinking water in india",size=25)
plt.legend(loc='upper left',fontsize=8)
plt.show()
```
